# creat a fake list of trusted informants 

```python
from faker import Faker
```

```python
fake = Faker()
```

```python
import csv
output=open('trusted_informants.csv', 'w')
fake=Faker()
header=['name', 'age', 'city']
mywriter=csv.writer(output)
mywriter.writerow(header)
for r in range(100):
    row=[fake.name(), fake.random_int(min=18, max=80, step=1), fake.city()]
    mywriter.writerow(row)

output.close()
```
